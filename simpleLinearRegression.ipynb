# Simple Linear Regression
>**tesk** : pradict the sudent score
>**input** : hours of studies
>**output** : score

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('w1_student_scores-220925-152435.csv')
```

```python
df.head(10)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (6,4))
plt.plot(df['Hours'],df['Scores'],'ro')
plt.title('student score vs hours of studies')
plt.xlabel('hours')
plt.ylabel('Scores')

plt.show()
```

## find coeffetietns

calculate the mean values of input and output

```python
xmean = np.mean(df['Hours'])
ymean = np.mean(df['Scores'])
```

calculate the numerator and denominater for beta

```python
df['xycov'] = (df['Hours'] - xmean) * (df['Scores'] - ymean);
df['xcov'] = (df['Hours'] - xmean)**2
```

```python
slop = df['xycov'].sum() / df['xcov'].sum()
intercept = ymean - (slop*xmean)
print(f'slop is: = {slop}')
print(f'intercept is : {intercept}')
```

## making pradiction

```python
predicted_score  = intercept + slop*df['Hours']
```

comparition between pradicated value and grouded data

```python
plt.plot(df['Hours'],predicted_score)
plt.plot(df['Hours'],df['Scores'],'ro')
plt.title('pradicted vs actual score graph');
plt.xlabel('hours');
plt.ylabel("score");
plt.show()
```
